  ```

  ```
  - 한글 검사 x, 잘 입력되었다고 가정
  - 사용자가 게임 제목을 입력하고, 엔터를 치면 게임 제목이 입력된다 => input()
  - 사용자가 입력 후 엔터를 칠때까지 무한대기 한다(언제 입력할지 모르니깐) => 무한루프 => while True

      - 공백만 넣고 엔터 => 공백제거(strip()) => 체크됨
      - 오류 메세지 : "정확하게 입력하세요"
    - 정확하게 입력한다면
      - stage2로 이동 (구현 x, stage1 완료)
      - 게임 제목을 받는 변수는 game_title로 설정
      - 무한 루프 종료

````python
while True:
  game_title = input('게임 제목을 입력...\n').strip()
  # 입력값 검사 -> 입력값은 문자열이다
  print(type(game_title), game_title)
  if len (game_title) == 0 : # 입력값이 없다 (사용자가 입력한 것이 없다)
    print('정확하게 입력하세요')
  elif len(game_title) > 28:
    print('28자 이내로 입력하세요')
  else: # 내용이 존재하고, 28자 이내
    print('정상입력')
    break
````

````python
# 실습 10분 : 14시 20분까지 최대한 구현해 보기

# 1-1 프럼프트 정리
#     글자 사이즈는 수치로 관리하겠금 조정 <= 컨셉 부여
GAME_TITLE_MAX_LENGTH = 28
game_start_prompt = f'''게임 제목을 입력해 주세요. (최대 {GAME_TITLE_MAX_LENGTH} 자이내(<={GAME_TITLE_MAX_LENGTH}), 반드시 영문 or 숫자 사용 가능)'
\n'''

while True:
  game_title = input(game_start_prompt).strip() # 입력값 획득 => 공백제거
  # 체크 진행
  # 입력 없이 엔터, 공백만 넣고 엔터 => 공백제거하면 => ''
  if not game_title: # if not '' : # 빈값을 잡아내는 조건식
    print('입력 없이 엔터, 공백만 넣고 엔터 상황 발생. 정확하게 입력하세요')
  elif len(game_title) > GAME_TITLE_MAX_LENGTH: # 28자 초과했는가?
    #print(f'{GAME_TITLE_MAX_LENGTH}자 초과. 정확하게 입력하세요')
    print(f'''
      게임 제목을 정확하게 입력하세요.
      방금 {GAME_TITLE_MAX_LENGTH}자 이내로 입력하시고,
      방금 입력한 타이틀은 {game_title}이고,
      최대 길이 대비 ({len(game_title) - GAME_TITLE_MAX_LENGTH})자가 초과 되었습니다.)
    ''')
  else: # 정확하게 입력한 케이스
  ##  print(f'입력하신 게임 타이틀은 {game_title}입니다.')
  ##  print('정상입력 되었습니다. stage2로 이동됩니다')
    break
````

# Stage 2

- 플레이어의 이름을 stage1 과 동일한 절차로 입력받는다
  - 제약 조건 => continue를 무조건 사용
- 변수명 : player_name

````python
while True:
  player_name = input('플레이어 이릉을 입력...\n').strip()
  print(type(player_name), player_name)

  if len (player_name) == 0 :
    print('1. 정확하게 입력하세요')
    # 반복문의 조건식으로 점프, 아래 코드 진행 x
    continue

  if len(player_name) > 28:
    print('2. 정확하게 입력하세요')
    # 반복문의 조건식으로 점프, 아래 코드 진행 x
    continue

  break
````

````python
# 2-1. 프럼프트 정리
PLAYER_NAME_MAX_LENGTH = GAME_TITLE_MAX_LENGTH # 28
player_name_prompt = f'''플레이어의 이름을 입력해 주세요. (최대 {PLAYER_NAME_MAX_LENGTH} 자이내(<={PLAYER_NAME_MAX_LENGTH}), 반드시 영문 or 숫자 사용 가능)'
\n'''

# 2-2. 무한루프, 사용자 입력이 정확하게 처리될때까지
while True:
  # 2-2-1. 프럼프트 출력 -> 입력 대기 -> 입력하고 엔터치면 -> 공백제거 -> player_name 세팅됨
  player_name = input(player_name_prompt).strip()

  # 2-2-2. 입력값 검사
  if not player_name: # 빈값 검사
    print('입력 없이 엔터, 공백만 넣고 엔터 상황 발생. 정확하게 입력하세요')
    continue # 반복문의 조건식으로 점프

  if len(player_name) > PLAYER_NAME_MAX_LENGTH: # 28자 초과 체크
    print(f'''
      플레이어의 이름을 정확하게 입력하세요.
      방금 입력한 이름은 {player_name}이고,
      최대 길이 대비 ({len(player_name) - PLAYER_NAME_MAX_LENGTH})자가 초과 되었습니다.)
    ''')
    continue # 반복문의 조건식으로 점프

  # else: # 정상입력
  break # 입력 대기 종료(반복문 종료)
````

````python
# 1회만 작동
ai_number = random.randint(1, 100)

# 아래 코드 전체는 정확하게 넣을때까지 무한 반복
while True:
  player_number = input('AI가 생성한 값 1~100 사이중 하나의 정수값을 맞추시오').strip()

  if not player_number:
    print('1. 정확하게 입력하세요')
    continue # 반복문의 조건식으로 점프
  elif player_number.isnumeric() == False:
    print('2. 정확하게 입력하세요')
    continue # 반복문의 조건식으로 점프
  else:
    # 3-3. 1 <= 입력 <= 100 에 들어오는지 검사
    # 입력값(문자열)을 정수(수치형)으로 변환
    p_num = int(player_number)
    # 크기 비교 가능
    # p_num 1보다 크거나 같'고', 100보다 작거나 같다
    # 고 => and, 거나 => or
    # if (p_num >= 1) and (p_num <= 100): # 정상적인 값 체크
    # p_num이 1보다 작거나, 100보다 크면 -> 오류
    if (p_num<1) or (100<p_num):
      print('3. 정확하게 입력하세요')
      continue # 반복문의 조건식으로 점프

  # 현재 코드는 정확하게 혹은 부정확하게 입력해도 무조건 아래 라인 수행됨
  # 조정 -> 잘못 입력되면 -> 다시 입력되게 처리
  # 아래 코드는 정확하게 입력했을때 진입하는 코드로 설정
  print(ai_number, player_number)
  # 이 부분이 수행된다 -> 정확하게 입력했다 -> 더이상 입력받을 필요 없다 -> 탈출
  break
````

# Stage 7

- 다시 게임을 하시겠습니까?
  - yes, Yes, Y, y, yEs, yeS => 모두 OK(긍정으로 인식)
    - stage 4로 진행
  - No, NO, no, nO, n, N => 모두 부정으로 인식
    - 게임 종료 -> 프로그램 종료
    - 'bye bye~'

````python
# 아래처럼 흐름을 컨트롤 할 수 있는 변수 => 플래그 변수라고 하고, 로직임(문법x)
is_game_playing = True # 게임 진행중인가?, 게임 그만하면 False가 된다
while is_game_playing: # <-> while True
  # stage 4
  ai_number = random.randint(1, 100)
  # print(ai_number) ## 편의를 위해 ai값을 보여줌
  try_count = 0

  while True:
    # stage 5
    while True:
      player_number = input('AI가 생성한 값 1~100 사이중 하나의 정수값을 맞추시오').strip()
      if not player_number:
        print('1. 정확하게 입력하세요')
        continue
      elif player_number.isnumeric() == False:
        print('2. 정확하게 입력하세요')
        continue
      else:
        p_num = int(player_number)
        if (p_num<1) or (100<p_num):
          print('3. 정확하게 입력하세요')
          continue

      break

    player_number = p_num

    # stage 6
    try_count += 1
    if player_number > ai_number:
      print('힌트 - 정답보다 크다')
      pass
    elif player_number < ai_number:
      print('힌트 - 정답보다 작다')
      pass
    else:
      print('정답')
      break

  # stage 7
  # 점수 출력
  print(f'{(11-try_count)*10}점 획득했습니다.')

  while True:
    # 대소문자 섞어서 표현, 대문자로 하던 관계없이 처리하겠다 -> 내부적으로 모두 소(대)문자 처리
    player_answer = input('다시 게임을 하시겠습니까? (yes|no, y|n)').strip().lower()
    if player_answer == 'yes' or player_answer == 'y':
      is_game_playing = True
      # stage 4 (새로 게임 시작 => AI 숫자부터 다시 생성)
      break
      pass
    elif player_answer == 'no' or player_answer == 'n':
      # 종료로 이동
      print('bye bye~')
      is_game_playing = False
      break
      pass
    else:
      print('정확하게 입력하세요')
````
