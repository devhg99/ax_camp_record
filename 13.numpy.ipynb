### arange()

- 파이썬 range()의 numpy 버전
  - 연속수를 생성 (제너레이터 기능 탑재: 참고)
  - 기능은 거의 동일

```python
# 0 <= x < 5 => 연속된 수를 가진 배열 생성함수
# step 간격으로 연속적
np.arange(5)
```

```python
# (시작인덱스, 끝인덱스, step (간격))
# 0 <= x < 5, step 2
np.arange(0,5,2)
```

```python
# 연속적인 배열 생성 -> reshape 같이 활용 -> 다차원 배열 연속수 생성
np.arange(1,17)
```

```python
# reshape() 형태를 변경
# 1차원 배열 => 2차원 배열 변경(차원변경)
np.arange(1,17).reshape(4,4)
```

```python
# 배열의 맴버수는 차원이 변경되어도 동일
# 계산이 힘든 경우도 존재함 -> -1를 넣으면 알아서 계산해줌
# 1차원이 8이므로, 2차원을 알아서 2로 계산해줌, 총수가 16이니깐
np.arange(1,17).reshape(8, -1)
```
