# 조건식

- 특징
  - 하위 주제의 전제 조건

- 결론
  - 조건식의 결론 => 블린형 데이터 (true, False)
  - 특징
    - False로 인식하는 표현
      - None,'', 0, 0.0, [], {}, (), False
      - 조건식의 결과가 False인 경우
- 연산자
  - 값 대 값 비교 조건식
    ```
      A == B : A와 B는 값이 일치하는가?
      A != B : A와 B는 다른가?
      A > B : A는 B보다 큰가?
      A < B : A는 B보다 작은가?
      A >= B : A는 B보다 큰거나 같은가?
      A <= B : A는 B보다 작거나 같은가?
      not A : A값을 부정한다
        - 만약 A값이 False면 -> 조건식 결과는 True -> 부정적 상황 잡아내는 조건식
    ```
  - 조건식 대 조건식 간 연산자(쇼트 연산자)
    - (조건식=C1) 연산자 (조건식=C2)
    - 조건이 복잡한 경우, 빠르게 처리하기 위함(쇼트연산)
    - C1 or C2
      - 한건체크 종료
      - C1이 **참**이면 그대로 조건 체크 종료

    - C1 and C2
      - C1이 거짓이면 그대로 조건 체크 종료
      - 한개의 조건식의 **거짓**이면 나머지는 체크할 필요 없음

  - 기타
    - A is B
      - A와 B가 동일한가? => 내부적으로 관리하는 id값을 비교
      - 동일 객체인지 (같은 대상을 참조하는지) 체크
    - a in A
      - A안에 맴버로 a가 존재하는가?

````python
# 고객의 체크카드 잔고(예산)
my_money = 5000

while True:
  coffe_price = int( input('커피 주문 총액을 입력하세요\n').strip() )
  if coffe_price == 0: ## (0에 커피값입력) -> break = 탈출
    break

  # 여기서 조건문 구현 -> 조건의 총 수는?
  if coffe_price > my_money:
    print(f'{coffe_price - my_money}원이 부족합니다.')
    pass # 자리를 차지하지만, 수행 안됨 => 틀 작업 시 사용
  elif coffe_price < my_money:
    print(f'잔액 {my_money - coffe_price}원이 남았습니다.') # x원 동적 계산되게 구현 및 오전 학습정리
    pass
  else: # 주문액과 잔고가 일치했음
    print('정상 결제되었습니다.')
    pass

print('bye bye~')

# 추가 요구사항 -> 오후 학습정리때 진행
# 결제한 금액만큼 잔액에서 제거(업데이트 진행) -> 잔고가 줄어든다 : 이 기능 반영
# 위의 요구사항의 결론 => 잔고 변수의 값을 업데이트(증(+)감(-))해라
````

````python
# 사본 셀에서 작업 -> 결제가 성공하면, 잔고 업데이트

# 고객의 체크카드 잔고(예)
my_money = 5000

while True:
  coffe_price = int( input('커피 주문 총액을 입력하세요\n').strip() )
  if coffe_price == 0: ## (0에 커피값입력) -> break = 탈출
    break

  if coffe_price > my_money:
    print(f'{coffe_price - my_money}원이 부족합니다.')
    pass
  elif coffe_price < my_money: # 잔고 변경
    # 잔고
    # 잔고 = 잔고 - 커피가격
    my_money = my_money - coffe_price
    # +=, -=, *=, /=, .. 모두가능 : 기존 변수에 특정 연산을 수행하고 다시 변수값으로 세팅할때
    # my_money -= coffe_price # 위의 코드와 동일, 통상 이렇게 표현, 기존값을 수정하여 다시 세팅
    print(f'잔액 {my_money}원이 남았습니다.') ## 업데이트 진행(수정)
    pass
  else: # 잔고 변경

    print('정상 결제되었습니다. 잔액은0원입니다.')
    pass

print('bye bye~')
````

### 리스트 + 튜플을 적용된 데이터에 반복문 처리

````python
# 데이터 형태 [ (,..), (,..), (,..), ...]
# 데이터는 리스트인데, 맴버는 튜플인 구조
datas = [(1,2), (3,4), (5,6)]
datas
````

````python
# 요구사항 출력형태는 다음과 같다
'''
첫번째값 1, 두번째값 2
첫번째값 3, 두번째값 4
첫번째값 5, 두번째값 6
'''
# 구현하시오
for data in datas:
  print(f'첫번째값 {data[0]},두번째값 {data[1]}')
````

````python
# 튜플 => 분해 작업이 같이 연상됨
for data in datas:
  a, b = data
  print(f'첫번째값 {a},두번째값 {b}')
````

````python
# 받으면서 바로 분해 or 분해해서 받는다
# (온도, 습도)라고  가정
for 온도, 습도 in datas: # 데이터를 명확하게 규정 가능함
  print(f'첫번째값 {온도},두번째값 {습도}')
````

````python
# 습도만 필요하다, 필요없는 값은 받고 버리기 => _
for _, 습도 in datas:
  print(f'습도 {습도}')
````

### 딕셔너리를 적용된 데이터에 반복문 처리


````python
subways = {
    'name':'서울역',
    'age':125
}
subways
````

````python
# 딕셔너리를 반복문에 기본 적용한다면? -> 키가 출력됨
for key in subways:
  print(key)
````

````python
# 키와 값(인덱싱 사용)을 같이 출력한다면?
for key in subways:
  print(key, subways[key])
````

````python
# 위 코드를 개선 -> 키와 값을 한번에 뽑는 코드
# subways.items() -> 딕셔너리의 items() 함수 -> (키,값) 페어해서 리스트로 반환
for key, value in subways.items():
  print(key, value)
````

##### enumerate(), 인덱스 부여

- enumerate(연속형데이터)
  - (인덱스, 딕셔너리의키|이스트값|튜플의값..) 반복적으로 반환


````python
# 순서가 없는 딕셔너리에 순서의 개념인 인덱스를 같이 추출하고 싶다면?
# 내장함수 enumerate()를 사용 -> 0, 1, 2, 3 ..같이 추출하고 싶다
for key in enumerate(subways):
  # 인덱스가 같이 추출되었음
  print(key)
````

````python
# 실습, 아래처럼 출력 되도록 코드를 작성하시오
'''
출력결과
인덱스 키 값
-----
0 name 서울역
1 age 125
'''
# 코드의 가독성이 떨어짐, 결과는 나옴
for subway in enumerate(subways):
  print(f'{subway[0]} {subway[1]} {subways[subway[1]]}')

# 데이터는 (인덱스, (키,값)) 이 형태로 추출됨
# 튜플이 2번 부분적으로 적용되어 있는 구조
for index, (key,value) in enumerate(subways.items()):
  print(index, key, value)
  # print(f'{index} {key} {value}')

# 결론 : 반복문의 결과로 나오는 데이터를 보고 판단(분핸, 직접 인덱싱, 그냥 사용...)
````

### range()

- 특정 범위의 수를 생성(?)하여 공급하는 함수

- 내장함수 : 바로 사용 가능
- 특정 범위내에서 **수치를 생산 후 제공(공급)**담당
- 제너레이터 기능이 탑재되어 있음
  - 필요한 수치를 미리 다 생성 x, 요청할때 만들어서 제공 -> **메모리를 효율적 사용**

- 사용
  ```
    range(시작값, 끝값, step)

    시작값(생략되면 0부터) <= x < 끝값
    step 1이 기본값
  ```

````python
# 요구사항 0, 1, 2 값을 생산하여 출력하시오
# range(3) => range(0, 3)
for x in range(3):
  print(x)

print(['-'*10])

# 요구사항 1<= x < 3
for x in range(1, 3):
  print(x)

print(['-'*10])

# 1 <= x < 10, 간격은 2 단위
for x in range(1, 10, 2):
  print(x)
````

### 이중for문

- for문 안에 for문이 존재 -> 중첩 반복문

    - 구구단 구현

````python
# n이 0일때 m은 0, 1, 2를 변화하면서 출력
# n이 1일때 m은 0, 1, 2를 변화하면서 출력
for n in range(2):   # 0 <= n < 2 : 0, 1
  for m in range(3): # 0 <= m < 3 : 0, 1, 2
    print(f'{n}x{m}={n*m}')
````

````python
# 구구단 구현, 3단, 4단, 5단만 구현
# 출력 형식
'''
  3 x 1 = 3
  3 x 2 = 6
  ...
  5 x 9 = 45

  - 추가 구현
  - 모든 문자는 2자리를 차지
  - 첫번째 숫자는 왼쪽 정렬, 두번째 숫자는 오른쪽 정렬, 세번째 숫자는 중앙정렬
'''
for i in range(3,6):    # row
  for j in range(1,10): # col
    print(f'{i:<2}x{j:2}={i*j:^2}')
````

## while문

- 언제 끝날지 모른다. 무한루프
- 0 ~ 무한대
- 반드시 탈출 코드가 필요함
  - case 1 : 조건 + break
  - case 2 : while문의 조건식을 통해서 종료
  - case 3 : 플래그 변수(로직으로 구성)를 통해서 값을 변경시켜서 종료
    - 중첩된 반복문이 3중 4중.. 겹쳐져있으면 유용함

- 문법

```
  while 조건식(T or F):
    statements
```

````python
# 더미 데이터, 리스트, 맴버 3개
datas = [11,12,13]

# 요구사항, datas에 맴버가 존재하면, <= 조건식
#           datas의 데이터를 하나씩 뒤에서부터 제거하여 출력하시오 <= 리스트.pop()
#           datas의 데이터가 없을때까지 반복하시오 <= 언제 없어지지?, 특정 x => while
print('반복문전 - 맴버수', len(datas))

while len(datas) > 0: # datas에 맴버가 존재하면
  print('A - 맴버수', len(datas))
  print('맴버수', len(datas), '제거 후 출력',datas.pop())
  print('B - 맴버수', len(datas))

print('반복문후 - 맴버수', len(datas))
````

````python
# 더미 데이터, 리스트, 맴버 3개
datas = [11,12,13]

# 조건식에 False 상황 ; 0, 0.0, [], (), {}, None, False, ...
while len(datas) > 0: # datas에 맴버가 존재하면
  # len(datas) 값은 언제나 0이 된다. 계속해서 datas.pop() 하고 있으므로
  print(datas.pop())
````

````python
# 더미 데이터, 리스트, 맴버 3개
datas = [11,12,13]

# 파이썬스러운 코드임
while datas: # [11,12,13] -> [11,12] -> [11] -> [] => False
  print(datas.pop())
````
