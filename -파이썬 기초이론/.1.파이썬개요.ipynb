## 프로그램 스타일

# 타입

- 정의
  - 데이터(수치(1, 2, 온도, 집값, 키, ..)
    - 문자열(리뷰, 댓글, ..)
    - 데이터를 담는 그릇이 프로그램 작성시 필요 -> 데이터를 연산등을 통해서 목적 달성(S.W) -> 그릇은 변수(variable) -> 변수는 타입을 가질 수 있다(규칙) -> 수치를 담는 그릇, 문자열을 담는 그릇...
  - 타입 -> 데이터의 형태를 규정한 것

- 종류 (러프하게 구성)
  - **단일 데이터 형**
    - 데이터 자체 본질 표현
    - **수치형**
      - 정수 : ..., -1, 0, 1, ..
      - 부동 소수: 3.14
      - 복소수
    - **문자열**
      - 'hello'
      - "안녕하세요"
    - **블린형**
      - True, False

  - **연속 데이터 형**
    - **단일 데이터가 여러개 모여서 구성**
      - 문자열은 단일데이터형 이면서 연속 데이터 형임
    - 종류  
      - 파이썬
        - list (리스트)
          - 순서가 있다
        - dict (딕셔너리)
          - 키와 값으로 데이터 관리
        - tuple (튜플)
          - 그냥 순서 있게 데이터를 묶는다
        - set (집합)
          - 중복된 데이터 제거하면서 관리
      
      - 데이터 분석 (데이터 사이언스, 어날리스트)
        - numpy 패키지의 ndarry(배열)
          - 수학 과학용 라이브러리
        - pandas 패키지의 Series/DataFrame
          - 데이터 분석용 라이브러리

      - 딥러닝
        - Tensor(텐서)
        - 직접 다루지는 않음(모델 생성 및 학습시 사용하는 용도)

  - **값이 없다**
    - **None**
  - 수치가 아니다. 결측치
    - NaN (numpy에서 표현)

- 시퀀스
  - 가변형 : list
  - 불변형(수정불가, readonly, immutable) : tuple, 문자열, byte(참고)

- 집합
  - set

- 매핑
  - dict
