## 맴버 함수

- 클레스내 함수

- 특별 함수(데코레이터 -> @xxx)

````python
# 맴바 함수 일반적인 형태
class Person:
  # 생성자
  def __init__(self, name):
    self.name = name # 인스턴스 변수
  # 맴버 함수
  def run(self):
    # 클레스 내부에서 맴버 접근 : self.맴버
    print(f'{self.name}run')
````

## 접근제어 (access control)

- 객체를 생성한 이후 특정 맴버를 접근할 수 없게 설정하는 기법
  - 은닉한다, 숨긴다 -> private
  - 클레스 내부에서만 사용하겠다

- 파이썬은
  - public -> 대부분 작성 코드
  - private -> 이름으로 설정, 키워드 x
    - __xxxx : 비공개, 노출 x, 외부접근 x
      - 객체명. <= 노출 x
    - _xxx : 뉘앙스를 표기, 내부용이다, 공개되지만 사용하지는 말아라

````python
class AccessControl:
  def __init__(self, _a):
    self._a = _a # 내부에서만 뉘앙스적으로 사용하겠다
    # 맴버 변수(인스턴스 변수)
    self.__a = 'hello' # 숨기겠다, private
    pass

  def __pri_func(self):
    print( '숨겨진 함수, 내부에서 사용', self._a, self.__a )

  def _pri_func(self):
    print( '뉘앙스적, 내부에 사용 하게다. 노출은됨', self._a, self.__a )
````

````python
ac = AccessControl('hi')

ac._a# ac.__a # 없다고 표현됨
````
