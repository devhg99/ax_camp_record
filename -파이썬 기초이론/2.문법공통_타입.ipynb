### 인덱싱

- 문자열에서 특정 문자열을 **추출**하는 행위
- 연속형 데이터형, 배열, 데이터프레임, ... 대부분 지원
- 본직적 의미 : **차원 축소**\
  - 문자열 예외
  - ex) 2D -> 1D (데이터의 구조가 줄어든다)
    - 리스트, 튜플, 딕셔너리, 배열, 시리즈, 데이터프레임, 텐서

- 문법
  - 인덱스의 방향성은 가까운쪽에서 해결

  ```
    변수명[ 정(역)방향인덱스 ]
  ```

````python
# 더미 데이터
a = 'abcdefghijklmn'
a
````

````python
# 요구사항 -> a라는 변수가 참조(담고있는)하는 데이터에 1번 자리(인덱스) 데이터 추출(인덱싱)
# 정방향 인덱스 : 0, 1, 2, ... <= 0부터 출발
# 요구사항 분석 -> 1번 자리는 두번째 데이터를 의미 -> 더미데이터에서는 "b"를 의미한다

# 인덱스는 커서의 위치, 데이터는 커서의 오른쪽값을 취한다 (1덩어리(청크)만 추출)
# 문자열에서는 문자 1개, 연속형 데이데이터는 데이터 한개(덩어리)
a[ 1 ]
````

````python
# 요구사항 -> 동일 결과를 역방향 인덱스로 표현하시오
# 역방향 인덱스 : -1, -2, -3, ... 이런 순으로 체크
# 요구사항 분석 -> 뒤에서부터 카운트 -> abcdefghijklmn -> 일일이 카운트 -> -13
#               -> 커서(특정 자리)에 오른쪽 값을 취한다
a[ -13 ]
````

- 정(역)방향 인덱스 적용은 원하는 데이터 가장 가까운 방향성 진행
- 데이터가 길면 -> 다른 로직을 적용해서 줄일 필요가 있다

### 슬라이싱

- 원본 데이터에서 특정 데이터를 자른다
- 차원 유지
- 2D -> 슬라이싱 -> 2D(백터)
- **연속적으로 이웃한** 형태의 데이터만 추출
  - step(간격)을 통해서 한칸 혹은 n칸 간격으로 추출가능함

- 문법
  - step 생략 가능 -> : 1개만 표현
  - 시작인덱스, 끝인덱스 값을 비울 수 있다 (처음부터, 끝까지)

  ```
    변수명[ 시작인덱스: 끝인덱스: step(간격,기본값1, 생략가능함)]
  ```

````python
a
````

````python
# 카피 => 슬라이싱 관점해석 => 처음부터 끝까지 자르시오
# 시작(끝) 인덱스 묘사 x => 처음(끝)부터 해석됨
a[ : ]
````

````python
# 해석 : 처음부터 인덱스 자리 2번까지 데이터 자른다(추출)
# 슬라이싱 적용 범위 -> 끝인덱스는 포함 안됨
# 시작인덱스 <= 인덱스 < 끝인덱스
# 0 <= 인덱스 < 2
# 결과값의 데이터 개수( 끝인덱스 - 시작인덱스, 2-0 => 2개 문자로 출력, 참고용)
a[ : 2 ]
````

````python
# -2 <= 인덱스 < 끝까지 => 맨뒤에서 순서대로 값 2개 추출
a[ -2:  ]
````

````python
# 요구사항 -> 원본 abcdefghijklmn -> 슬라이싱 -> bcdefghijklm
# 구현하시오, 2분
# 1<= 인덱스 < -1
a[1:-1], a[1:-1: 1]# step의 기본값이 1이므로 생략 가능
````

````python
# step이 2 이므로 하나씩 건너 뛰어서(간격 2) 추출
a[1:-1: 2]
````

### 문자열 주요 처리 함수(특정 기능을 가진 단위)

#### 분해

- 문자열을 특정 기준(구분자)으로 나눈다(통상 리스트([])로 표현됨)
- split()

````python
# 더미데이터 준비
url = 'https://m.sports.naver.com/wfootball/article/311/0001845199?id=123&aid=990'

# 요구사항
# url 텍스트에서 id값 추출하시오, id값, aid값, 각종 수치값들은 크기가 유동적임(참고)
# url의 id값 123을 추출하시오
# 요구사항 분석 -> ?를 중심으로 분해후 뒷값 선택, ... 반복적 수행 -> 최소범위로 텍스트가 추출됨
````

### (*)포멧팅

- 통상적으로 단일 데이터를 특정 문자열의 내부, 특정 형태로 구성하여 동적으로 문자열을 생성할때 유용

#### 첫번째 유형 - '문자열' % (값, 값, ...)

````python
x = 1
y = 2

# 요구사항 : x, y 변수를 이용하여 최종 출력 형식 '1 + 2 = 3' 형식으로 출력하시오
'%d + %d = %d' % (x, y, x+y)
````

````python
y = 1.1

# %d는 정수 표현, 데이터가 float => 데이터 손실 발생 => 적합한 형식 사용
'%d + %d = %d' % (x, y, x+y)
'%d + %f = %d' % (x, y, x+y) # 결과는 나오지만, 최종 형태는 아님
````

````python
# 쉬운 표현, 문자열로 형식을 %s 구성, 정보손실 없음, 자리수 일치
'%s + %s = %s' % (x, y, x+y)
````

#### 두번째 유형 - '문자열'.format(값, 값, ...)

````python
# 함수의 매개변수 순서대로 앞에부터 채워진다
'{} + {} = {}'.format(x, y, x+y)
````

````python
# 매개변수(함수 호출시 전달하는 데이터)를 다른 순서대로 배치하고 싶다면
'{0} + {1} = {2}'.format(x, y, x+y)

# 매개변수 순서를 다르게 적용하고 싶다면
'{1} + {0} = {2}'.format(x, y, x+y)
````

````python
# 매개변수를 전부 사용 x, 특정 매개변수는 2번 사용(중복)
'{1} + {0} = {0}'.format(x, y, x+y)
````

````python
# 의미를 명확하게 부여한다면? => 매개변수에 이름 부여
'{온도} + {습도} = {최종상태}'.format( 온도=x, 습도=y, 최종상태=x+y)
````

### 교집합, 차집합, 합집합

- 데이터베이스의 sql => join과 유사
- pandas => merge와 유사

````python
# 데이터
a1 = [1, 3, 4, 6, 2, 7, 8, 8, 3, 2, 100]
a2 = [5, 3, 6, 9, 256]
````

````python
# 중복제거
a1_set = set(a1)
a2_set = set(a2)

a1_set,a2_set
````

````python
# 교집합 => 2개의 집합에서 일치되는 데이터만 추출
a1_set.intersection(a2_set)
````

````python
# 합집합 => 2개의 집합을 모두 합쳐서, 중복제거
# sql => union
a1_set.union(a2_set)
````

````python
print( 'a1_set',a1_set )
# 차집합 => 오로지 a1_set에만 존재하는 데이터만 추출(고유함)
a1_set.difference(a2_set)
````
