# 모듈화/패키지

- 내가 만든 모듈을 다른사람이 사용하고 싶다면?
  - 팀내에서 A가 만든 기능을 B가 사용해야 한다면
  - A가 개발해서 공유, 제공 등등 -> B는 모듈 가져오기로 사용

```
# 구조
/ 현재 디렉토리
6. 모...ipynb
a         : 패키지
 L b      : 패키지
    L __init__.py : 모듈, 특정 디렉토리(패키지) 밑에 존재하면 해당 패키지를 대표(대변)하는 모듈로 작동 -> b 패키지
    L mod.py      : 모듈
 L__init__.py     : 모듈 -> a 패키지

```

- a와 b는 패키지임

# 예외처리

- 파이썬 진영 가이드
  - https://peps.python.org/pep-0463/

- 개요
  - 프로그램은 오류 발생시 셧다운 되면 안된다
    - 개발자 관점에서 이 부분 해결
      - 꼼꼼하게 체크 -> 오류 상황을 예측 미연에 방지
        - 조건문을 많이 사용하게 된다
        - LBYL : Look Before You Leap
          - 에러가 나올만한 상황을 꼼꼼하게 체크하여 구성
      - 코드 진행 -> 에러발생 -> 수습해(예외처리)
        - 코드는 쭉 진행, 수습은 나중에(개발단계)
        - EAFP : It's Easier to Ask Forgiveness than Permission
          - 일단 수행, 예외처리로 오류 수습(나중에 수습되는 스타일)
          - 파이썬 이 방식 권장함 -> 코드는 간결해짐
  - 결론
    - 회사 문화를 따라간다

- 예외처리
  - **오류**를 책입지고 **잡아서 처리**
    - try ~ except ~ else ~ finally ~
    - try ~ except ~ finally ~
    - try ~ except ~
    - try ~ except ~ else ~
  - 오류를 던지기
    - 한곳에서 모두 해결해라
    - raise ~

- 문법
```
  try:
    
    # 예외 상황이 발생될만한 코드, 의심스런 코드, I/O

  except 오류객체명1:

    # 오류 처리(로그등 처리)

  except 오류객체명2:

    # 오류 처리(로그등 처리)

  ...

  [else:] <- 생략 가능

    # 오류가 없으면 진입, 오류가 없었음을 알고 싶다면 사용

  [finally:] <- 생략 가능

    # 뒷정리 코드
    # 오류가 있던, 없던 간에 동일하게 처리할 내용 기술

 ```

- 오류객체
  - 최상위 객체(정점, 근본) Exception
  - 디테일한 오류 객체명을 모르면, Exception 사용
  - Exception은 모든 오류의 시초임

````python
# 예외처리문(:)
try:
  print(1)
  # print(1/0)
  print(2)
except Exception as e:
  print(3, e)
else:
  print(4, '오류없음')
finally:
  print(5, '뒷정리')
````

````python
# 예외 던지기 => 상위 코드에서 해결
def except_call():
  try:
    1/0
  except Exception as e:
    # 오류 던지기
    raise e


try:
  # 해당 함수에 오류가 발생했는데, 함수내에서 해결 x, 상위 코드에서 해결하게 처리한 예제
  except_call()
except Exception as e:
  print('오류처리', e)
````
