```python
# 실습2, df.payment_date에서 월정보 일괄 추출
# Series에서 데이터 하나씩 꺼내서 -> 전처리(월정보 추출) -> 다시 반환...
# apply()
# 데이터의 타입
# df.payment_date => 타입
# DF에서 컬럼 1개 선택(추출) -> 차원축소 -> Series
# type(df.payment_date)
df.payment_date.apply(lambda x: x[5:5+2]) # x = '2019-07-31 23:41:38'
# 02~07 : 7개월
```

```python
# 파생변수, df에 결제 정보중 월정보 별도 표현
df['pay_month'] = df.payment_date.c( lambda x: x[5:5+2] )

df.head(1)
```

```python
# apply() -> str로 변경하여 동일한 결과가 나오게 처리
# 실습
# 데이터가 많아지고, 년정보에땋라 02가 여러번 등장한다면 구분값 필요
# 월 => 년월로 변경!!
df['pay_month'] = df.payment_date.str[:7].str.replace('-', '')


df.head(1)
```
