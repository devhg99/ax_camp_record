# 워크플로우

## 요구사항 정의(연구 목표 수립)

- 구성
  - 타겟
    - 누구를 위한 서비스/보고서/분석...
    - 대상을 명확하게 설정
  - 목적(목표)
    - 분석
      - 설득, 의사결정의 판단지표, 마켓팅, ... => 데이터 애널리스트
    - 모델
      - 정량적 목표(기존대비 정확도 5% 상승시킨다 -> 프로젝트 종료가됨)
        - 기존 모델의 정확도 90%인데 95%까지 올린다
      - 신규서비스
        - 이커머스 : 추천, 고객분류, ... => 기존 유사 서비스등에 비교
  - 참고
    - 근거 제시(주장에 대한) 통계자료, 유사 데이터, 서비스 등 제시
  - 기타 필요 사항, ...
  - 마지막 단계를 염두하고 진행
    - **보고서, 논문, 서비스, 개선된 기능, ...**

## 데이터 수집(획득)

- 특징
  - 모든 과정은 **데이터를 기반**으로 진행됨
  - 초기 데이터가 없어도 가능한 방식 : 강화학습

- 수집 전략(난이도 기반 분류)
  - Level 1
    - 제공받는 데이터
      - 사내 데이터
        - 사내 타 부서에서 축적된 데이터
        - 업무 협조를 통해서 획득
      - 공공/대학/연구소/ 데이터
        - 공익을 위해 공개된 데이터
        - 요청 후. 절차 후 획득
        - 가공된 데이터
          - 국내 : 코로나때 공공 사업(데이터 구축 사업 진행)
        - AI 허브, 언어정보나눔터, ...
        - 각 지자체/국가행정기관에서 제공하는 데이터 포털
      - 바우처 사업, 데이터 거래소



---

- 수집 전략(난이도 기반 분류)
  - 공통
    - 인터넷을 통해서 획득
    - 상업적 사용시 법률 검토 필요
  
  - Level 2
    - open api 사용
    - ex)
      - 네이버, 카카오, 구글 등 => 가입 => **api 키 신청** => [rest api] 요청을 통해서 데이터 획득
      - 일일 쿼터 존재함
    - 사용 라이브러리
      - request, urllib : 기본 라이브러리
      - ...
    - 획득한 데이터는 포멧
      - xml, **json**

  - Level 3
    - web scraping (웹 스크레핑)
    - 조건
      - 웹 사이트 접속 -> 사용자 액션 x
      - OPEN API x
    - 절차
      - 웹 접속 -> 화면 구성 -> 화면의 소스(HTML)를 긁어(가져) 온다
      - 파이썬에서 HTML을 엑세스 -> 정보추출
        - HTML 탐색, 추출 등 행위 : css selector
        - 파이썬에서 css selector를 이용 정보 추출 라이브러리 : bs4
        - 단계
          - 페이지 요청 -> 응답 -> HTML -> 메모리 로드 : 파싱 -> 메모리 DOM(Document Object Model)이 로드 -> css selector 사용가능 -> 탐색 -> 데이터 추출
    - 필요 라이브러리
     - bs4
     - (*)css selector (필요기술), xpath, ...

  - Level 4
    - web crawing ([웹] 크롤링)
    - 조건
      - 원하는 데이터까지 진입하기 위해 필요한 사이트내 작업이 존재하면
        - 로그인
        - 검색
        - 스크롤 -> 데이터 더보기(불러오기)
        - 방화벽 조치 -> 10초 후에 화면이 구성된다(클라우드페어 사용)
      - **사람이 뭔가 수행하는 작업이 존재하는 경우**
    - 라이브러리
      - selenium
    - 사전에 이해하면 좋은 부분
      - 웹에 이해(필요한 만큼)
        - 프런트 구성
          - html : 구조, 데이터 가지고 있음
          - css : 레이아웃, 디자인
            - css selector
          - javasscript : 인터렉션(사람 <-> 브라우저), 이벤트
            - ajax 통신의 개요라도 이해
              - 검색등을 사용
    - 특징
      - 코랩 진행 x
      - 로컬 pc에서 환경 세팅
        - 파이썬 설치, ide 설치
      - 매크로 등이 여기에 해당됨(참고)


- 자동화
  - 알아서 특정 시간이 되면 프로그램이 자동 가동 -> 수집 -> (전처리옵션) 적제 (파일, 디비)
  - os 레벨
    - 윈도우 : 작업 스케줄러
    - 맥/리눅스 : cron 명령어

- 적제
  - 수집완료 이후 데이터 준비 단계쪽으로 데이터 전달을 어떻게 할것인지?
    - 정형 데이터
      - **데이터베이스, 엑셀/CSV**
    - 반정형 데이터
      - **JSON**, XML 구성
    - 비정형 데이터
      - 비구조적 텍스트

- 주의사항
  - 지속적 수집시
    - 문제점
      - 동일 ip에서 수집 -> 패턴이 읽힘(보안 관제 적용된 사이트)
        - 트레픽이 높으면 => ddos로 인식할 수 있음 => 차단됨
    - 해결
      - 수집 빈도 (주기), 랜덤 처리 등 조절
      - selenium 지원
        - 프록시를 통한 우회 접속
      - User_Agent 헤더값(조작->OS, 모바일로 진입 등)
      - ip 해결 => LTE를 사용(안드로이드폰으로 태더링, 온/오프 => 기지국 변경. IP바뀜)

  - ...

    - ...

## 데이터 분석

- 주요 기법
  - 통상적 머신러닝/딥러닝 까지 공통 적용되는 부분, 통계적인 약간배제
  - 설명분석
    - 기초 통계량, 데이터 특성확인
    - 데이터를 파악
  - 탐색분석
    - EDA, 시각화 -> 직관, 설명, 특징 파악
    - 이 데이터가 어떤 인사이트를 가지는지 파악
      - 변수들간에 잠재적인 관계 (인사이트, 숨겨진 의미 등) 해석
      - **주제가 인사이트 도출**

- 기타분석(통계관점)
  - 추론 분석
  - 예측 분석
    - 머신러닝/딥러닝 => AI
  - 인과 분석
  - 규범 분석

# 데이터 종류

- 수집/적제 전략 영향
- 분석/모델에도 영향을 미침

## 정형

- 구조화된 데이터
- 특징
  - 관계형 데이터베이스 (**RDB**) 기반으로 추출된 데이터
    - Oracle, Mysql, MS-SQL, ...
    - 데이터와 구조정보(스키마)분리되어서 구성
    - sql을 통해서 추출
  - open api를 통해서 획득한 데이터


## 비정형

- 구조가 없는 데이터
- 특징
  - **바이너리 데이터**
    - 영상, 이미지, 오디오, 문서(DOC, HWP, ...), ...
    - 해당 데이터를 다루는 파트 => 딥러닝
  - **Nosql** 계열 (<-> RDB)
    - 몽고디비, 레디스
    - 로그데이터, ...

## 반정형

- 데이터 안에 구조 정보(스키마)가 들어있는 데이터
- 특징
  - 종류
    - **json, xml**
      - 텍스트 기반, 이 기종/플렛폼/언어에 독립적인 데이터 포멧
      - 웹환경에서 가장 많이 주고받은 데이터 형식
    - html
    - ...
